```python
import numpy as np
import pandas as pd

import networkx as nx

import matplotlib.pyplot as plt
from pylab import rcParams
import seaborn as sb
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 5, 4
sb.set_style('whitegrid')
```

```python
DG = nx.gn_graph(7, seed=25)
# generate edgelist
for line in nx.generate_edgelist(DG, data=False): print(line)
```

```python
print (DG.node[0])
```

```python
DG.node[0]['name'] = 'Alice'
```

```python
print (DG.node[0])
```

```python
DG.node[1]['name'] = 'Bob'
DG.node[2]['name'] = 'Claire'
DG.node[3]['name'] = 'Dennis'
DG.node[4]['name'] = 'Esther'
DG.node[5]['name'] = 'Frank'
DG.node[6]['name'] = 'George'
```

```python
# add list of ages
DG.add_nodes_from([(0,{'age':25}),(1,{'age':31}),(2,{'age':18}),(3,{'age':47}),(4,{'age':22}),
                   (5,{'age':23}),(6,{'age':50})])
print (DG.node[0])
```

```python
DG.node[0]['gender'] = 'f'
DG.node[1]['gender'] = 'm'
DG.node[2]['gender'] = 'f'
DG.node[3]['gender'] = 'm'
DG.node[4]['gender'] = 'f'
DG.node[5]['gender'] = 'm'
DG.node[6]['gender'] = 'm'
```

```python
DG.node[0]
```

```python
# visualize network
nx.draw_circular(DG, node_color='bisque',with_labels=True)
```

```python
labeldict = {0: 'Alice',1:'Bob',2:'Claire',3:'Dennis',4:'Esther',5:'Frank',6:'George'}

nx.draw_circular(DG, labels=labeldict, node_color='bisque',with_labels=True)
```

```python
G = DG.to_directed()
```

```python
nx.draw_spectral(G, labels=labeldict, with_labels = True)
```
